# Objective 3
Visualize the data and summarize findings
Your final objective is to build visuals to show the number of new hosts by year, overall average price by year and neighborhood, and average price for various types of listings in Paris' most expensive neighborhood.

Create a horizontal bar chart of the average price by neighborhood in Paris, and make sure to add a title and change axis labels as needed


Create a horizontal bar chart of the average price by ‘accommodates’ in Paris’ most expensive neighborhood, and make sure to add a title and change axis labels as needed


Create two line charts: one showing the count of new hosts over time, and one showing average price. Set the y-axis limit to 0, add a title, and change axis labels as needed


Based on your findings, what insights do you have about the impact of the 2015 regulations on new hosts and prices?


BONUS: Create a dual axis line chart to show both new hosts and average price over time

```python
import matplotlib.pyplot as plt

paris_data = df[df['city'] == 'Paris']
avg_price_by_neighbourhood = paris_data.groupby('neighbourhood')['price'].mean().sort_values()

plt.figure(figsize=(10, 8))
avg_price_by_neighbourhood.plot(kind='barh', color='skyblue')
plt.title('Average Price by Neighbourhood in Paris')
plt.xlabel('Average Price')
plt.ylabel('Neighbourhood')
plt.show()
```

```python
most_expensive_neighbourhood = paris_data.groupby('neighbourhood')['price'].mean().idxmax()

expensive_neighbourhood_data = paris_data[paris_data['neighbourhood'] == most_expensive_neighbourhood]

avg_price_by_accommodates = expensive_neighbourhood_data.groupby('accommodates')['price'].mean().sort_values()

plt.figure(figsize=(10, 6))
avg_price_by_accommodates.plot(kind='barh', color='coral')
plt.title(f'Average Price by Accommodates in {most_expensive_neighbourhood}')
plt.xlabel('Average Price')
plt.ylabel('Number of Accommodates')
plt.show()
```

```python
df['host_since'] = pd.to_datetime(df['host_since'])

new_hosts_over_time = df.groupby('host_since')['host_id'].nunique()

average_price_over_time = df.groupby('host_since')['price'].mean()

plt.figure(figsize=(14, 6))

plt.subplot(1, 2, 1)
plt.plot(new_hosts_over_time.index, new_hosts_over_time.values, marker='o', linestyle='-', color='blue')
plt.title('Count of New Hosts Over Time')
plt.xlabel('Date')
plt.ylabel('Number of New Hosts')
plt.ylim(0)


plt.subplot(1, 2, 2)
plt.plot(average_price_over_time.index, average_price_over_time.values, marker='o', linestyle='-', color='green')
plt.title('Average Price Over Time')
plt.xlabel('Date')
plt.ylabel('Average Price')
plt.ylim(0)  

plt.tight_layout()
plt.show()
```
